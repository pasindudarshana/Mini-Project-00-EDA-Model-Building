**Churn Rate Calculation**

The churn rate is the proportion of customers who have left the company:

From the dataset:

- Number of customers who churned (`Yes`) = 1,869  
- Total customers = 5,174 (`No`) + 1,869 (`Yes`) = 7,043  

This means about **1 in 4 customers are leaving** the company.

---
**Class Imbalance and Implications**

- **Class Imbalance:**  
  - `No` (stayed) = 5,174 → 73.5%  
  - `Yes` (churned) = 1,869 → 26.5%  
  - The dataset is **imbalanced**, with the majority of customers staying.

- **Implications for Modeling:**  
  1. **Bias in Models:** Standard classifiers may favor the majority class (`No`) and misclassify churners (`Yes`).  
  2. **Evaluation Metrics:** Accuracy alone will be misleading. Metrics like precision, recall, F1-score, and ROC-AUC are more appropriate.  

**Summary:** About 26.5% of customers churn, indicating a moderate churn problem. Class imbalance must be considered during predictive modeling to avoid biased results.

#### 1.3.1 Numerical Features: Distribution analysis, outlier detection using IQR and Z-score methods  

**Z method**

```python
def find_anomaly(data):
    
    mean = data.mean()
    std = data.std()

    upper_bound = mean + 3 * std
    lower_bound = mean - 3 * std
    
    return (data > upper_bound) | (data < lower_bound)
```

```python
numerical_columns = [
                    'tenure', 'MonthlyCharges',
					'TotalCharges'
                    ]

for col in numerical_columns:
    n_outliers = find_anomaly(df[col]).sum()
    n_outlier_percentage = round(n_outliers / len(df), 3)
    print(f"{col} : {n_outliers} outliers ({n_outlier_percentage}% )")
```

**IQR Method**

```python
def find_anomaly(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    return (data > upper_bound) | (data < lower_bound)

```

```python
numerical_columns = [
                    'tenure', 'MonthlyCharges',
					'TotalCharges'
                    ]

for col in numerical_columns:
    n_outliers = find_anomaly(df[col]).sum()
    n_outlier_percentage = round(n_outliers / len(df), 3)
    print(f"{col} : {n_outliers} outliers ({n_outlier_percentage} %)")
```

```python
outlier_mask_log = pd.DataFrame()
for col in numerical_columns:
    outlier_mask_log[col] = find_anomaly(df[col])

outlier_count = outlier_mask_log.sum(axis=1)
outlier_mask_log['Outlier Count'] = outlier_count
outlier_mask_log = outlier_mask_log[outlier_mask_log['Outlier Count'] > 1]
outlier_mask_log
```

#### 1.4.4 Statistical Significance: Use appropriate tests (Chi-square, t-tests) to validate relationships

```python
from scipy.stats import chi2_contingency


# Example: InternetService vs Churn
contingency_table = pd.crosstab(df['InternetService'], df['Churn'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square Test: InternetService vs Churn")
print(f"Chi2 = {chi2:.3f}, p-value = {p:.5f}")
```

```python

# # Separate categorical and numerical features
# categorical_features = df.select_dtypes(include=['object']).columns.drop('Churn')
# numerical_features = df.select_dtypes(include=['int64', 'float64']).columns

results = []

# Chi-square test for categorical features
for col in categorical_columns:
    contingency_table = pd.crosstab(df[col], df['Churn'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    results.append({
        'Feature': col,
        'Test': 'Chi-square',
        'Statistic': chi2,
        'p-value': p
    })

# t-test for numerical features
for col in numerical_columns:
    churn_yes = df[df['Churn'] == 'Yes'][col].dropna()
    churn_no = df[df['Churn'] == 'No'][col].dropna()
    t_stat, p = ttest_ind(churn_yes, churn_no, equal_var=False)
    results.append({
        'Feature': col,
        'Test': 't-test',
        'Statistic': t_stat,
        'p-value': p
    })

# Create results table
results_df = pd.DataFrame(results)
results_df['Significant (p<0.05)'] = results_df['p-value'] < 0.05

results_df
```

**Non Significant Features** are gender and Phone Service.
